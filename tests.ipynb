```python
# Import the package
from isotopia import RadioactiveDecay, TracerModel, Plotter
import numpy as np

# Define time intervals
time_intervals = np.linspace(0, 10, 100)

# Test 1: Radioactive Decay Calculator
decay = RadioactiveDecay(decay_constant=0.1)
remaining = decay.calculate_remaining(initial_amount=100, time=time_intervals)
print(f"Half-life: {decay.calculate_half_life():.2f} units")

# Test 2: Tracer Concentration Over Time
tracer = TracerModel(decay_constant=0.1, dilution_rate=0.05)
concentration = tracer.concentration_over_time(initial_concentration=50, time=time_intervals)

# Test 3: Plotting
Plotter.plot_concentration(time_intervals, remaining, "Remaining Isotope Over Time")
Plotter.plot_concentration(time_intervals, concentration, "Tracer Concentration Over Time")
```
